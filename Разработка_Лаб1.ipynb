#Задание

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def shewhart_chart(data, client_name):
    mean = np.mean(data['Сумма'].astype(float))
    std_dev = np.std(data['Сумма'].astype(float))

    ucl = mean + 3 * std_dev
    lcl = max(0, mean - 3 * std_dev)

    plt.figure(figsize=(10, 6))
    plt.plot(data['Дата'], data['Сумма'], marker='o', linestyle='-', label='Сумма')
    plt.axhline(mean, color='green', linestyle='--', label='Среднее')
    plt.axhline(ucl, color='red', linestyle='--', label='UCL (УКГ)')
    plt.axhline(lcl, color='red', linestyle='--', label='LCL (НКГ)')
    plt.title(f'Карта Шухарта для {client_name}')
    plt.xlabel('Дата')
    plt.ylabel('Сумма')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return mean, ucl, lcl

def range_chart(data, client_name):
    
    data['Размах'] = data['Сумма'].diff().abs()

    r_bar = np.mean(data['Размах'].dropna())  
    ucl_r = 3.267 * r_bar  
    lcl_r = 0  
    
    plt.figure(figsize=(10, 6))
    plt.plot(data['Дата'], data['Размах'], marker='o', linestyle='-', label='Размах')
    plt.axhline(r_bar, color='green', linestyle='--', label='Средний размах')
    plt.axhline(ucl_r, color='red', linestyle='--', label='UCL (Верхняя контрольная граница)')
    plt.axhline(lcl_r, color='red', linestyle='--', label='LCL (Нижняя контрольная граница)')
    plt.title(f'R-карта для {client_name}')
    plt.xlabel('Дата')
    plt.ylabel('Размах')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return r_bar, ucl_r, lcl_r

def analyze_clients(clients_dfs):
    clients_analysis = []

    for i, df in enumerate(clients_dfs, start=1):
        mean, ucl, lcl = shewhart_chart(df, f'Клиент {i}')

        r_bar, ucl_r, lcl_r = range_chart(df, f'Клиент {i}')

        if (df['Сумма'].astype(float) > ucl).any():
            category = 'Растущие'
        elif (df['Сумма'].astype(float) < lcl).any():
            category = 'Проблемные'
        else:
            category = 'Лояльные'

        clients_analysis.append((f'Клиент {i}', mean, ucl, lcl, r_bar, ucl_r, lcl_r, category))

    return pd.DataFrame(clients_analysis, columns=['Клиент', 'Среднее', 'UCL', 'LCL', 'Средний размах', 'UCL размах', 'LCL размах', 'Категория'])

clients_dfs = [df1, df2, df3, df4, df5, df6]
clients_summary = analyze_clients(clients_dfs)

print(clients_summary)
```
